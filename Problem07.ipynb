```python
#QUESTION01
#DESIGN CLASS
class Student():
    
    id_number = 0
    
    def __init__(self, name, department, age, cgpa):
        Student.id_number += 1
        self.id = Student.id_number
        self.name = name
        self.department = department
        self.age = age
        self.cgpa = cgpa
    
    def get_details(self):
        print("ID: " + str(self.id))
        print ("Name: " + self.name)
        print ("Department: " + self.department) 
        print ("Age: " +  str(self.age))
        print ("CGPA: " + str(self.cgpa))

    #CLASS METHOD
    @classmethod
    def from_String (cls, var1):
        attributes_list = var1.split("-")
        return (cls(attributes_list[0], attributes_list[1], attributes_list[2], attributes_list[3]))

#MAIN CODE
s1 = Student("Samin", "CSE", 21, 3.91)
s1.get_details()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.get_details()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.get_details()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.get_details()

#SUBTASK05
print ("A class variable retains its assigned value and can be changed throughout the entire set of program codes. However, the instance variable only exists within an object and can be changed only within that particular object.") 

#SUBTASK06
print ("A class method is involved within a class which runs throughout the class. However, an instance method only exists in an object and runs within that particular object.")
```

```python
#QUESTION03
#DESIGN CLASS
class Dates():
    def __init__(self, date):
        self.date = date
        
    def getDate(self):
        return (self.date)
    
    @classmethod
    def toDashDate(cls, date):
        var1 = date.split ("/")
        var2 = "-"
        dashDate = var2.join(var1)
        return (cls(dashDate))

#MAIN CODE
date1 = Dates("05-09-2020")
dateFromDB = "05/09/2020"
date2= Dates.toDashDate(dateFromDB)
if(date1.getDate() == date2.getDate() ):
    print("Equal")
else:
    print("Unequal")
    
#SUBTASK04
print ("The conditional statement prints equal because as we replace given date format to the required date format, the string we obtain for the required date format is the same as the string of the given date format. The string of the given date format acts as the parameter in the following class. As a result, when we finally call the getDate() function, it provides us with the values which were stored in variables instead of their memory locations. Hence, they are the same.") 
```

```python
#QUESTION04 
#DESIGN CLASS
import math
class Circle ():
    def __init__ (self, radius):
        self.__radius = radius
        self.__area = math.pi * (self.__radius**2)
        
    def getRadius(self):
        return (self.__radius)
    
    @classmethod
    def setRadius (cls,var1):
        return (cls(var1))
    
    def area (self):
        return (self.__area)
    
    def __add__ (self,var1):
        return (Circle.setRadius(self.__radius + var1.__radius))
    
#MAIN CODE
c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" ,c1.area())
c2 = Circle(5)
print("Second circle radius:" ,c2.getRadius())
print("Second circle area:" ,c2.area())
c3 = c1 + c2
print("Third circle radius:" ,c3.getRadius())
print("Third circle area:" ,c3.area())
```
